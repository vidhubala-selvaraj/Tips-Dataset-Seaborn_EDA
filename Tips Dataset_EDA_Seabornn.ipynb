```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names() #list of datasets for practice
```

```python
tips = sns.load_dataset('tips') #loading dataset 'tips'
print(tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False) #sns.lmplot(x = 'column_name', y = 'column_name', data = data_set_name)
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'day', data = tips, fit_reg = False) # hue says w.r.t
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips, fit_reg = False) # hue says w.r.t, it is the 3rd sub-attribute
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'sex', data = tips, fit_reg = False) # 'col' says w.r.t unique values in the category
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'day', data = tips, fit_reg = False)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'day', col = 'sex', row = 'time', data = tips, fit_reg = False)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'day', palette = 'spring', data = tips, fit_reg = False) #palatte = '__' - gives different shades
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'day', palette = 'magma', data = tips, fit_reg = False)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', palette = 'summer', markers = ['x', 'o'], data = tips, fit_reg = False)
plt.savefig('My_Plot_1.jpg') #saves figure in the location where the current notepad was opened
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = False) #adv. - better described than bar plot, disadv. - overlapping
plt.show()
```

```python
x = tips['day']
y = tips['total_bill']
plt.bar(x , y)
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips) # does not overlap points - adv.
plt.show()
```

```python
sns.countplot(tips['sex']) # or can be written as sns.countplot(x = 'day', data = tips) - gives count
plt.show()
```

```python
sns.countplot(x = 'smoker', hue = 'sex', data = tips)
plt.show()
```

```python
sns.distplot(tips['total_bill']) #shows the graph
plt.show()
```

```python
sns.distplot(tips['tip'])
plt.show()
```

```python
tips['total_bill'].skew() #gives skewness value
```

```python
tips['tip'].skew()
```

```python
tips['total_bill'].kurt() #kurtosis value
```

```python
#pearsonr tells about the distribution and correlation where pearsonr value > 0.5 is decent
#bi-variate analysis - pearsonr and univariate analysis - histogram
sns.jointplot(x = 'total_bill', y = 'tip', data = tips) #jointplot gives both hist and scatter distribution.
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex') #kind = hex gives hex kind of shapes of points
plt.show()
```

```python
sns.boxplot(tips['total_bill']) #univariate
plt.show()
```

```python
sns.boxplot(tips['total_bill'], tips['smoker'], hue = tips['sex']) #univariate
plt.show()
```

```python
sns.boxplot(tips['day'], tips['total_bill'], hue = tips['sex'])
plt.show()
```
